```python
import pandas as pd
import plotly.express as px
```

```python
info_carros=pd.read_csv("C:/Users/gabri/personal_works/vehicles/vehicles_us.csv")
info_carros
```

```python

price_for_cylinders=info_carros.pivot_table(index="cylinders", values ="price", aggfunc="mean").reset_index()
price_for_cylinders["price"]= price_for_cylinders["price"].round()
```

```python
price_for_cylinders.plot(kind='bar', x="cylinders", y="price")
```

```python
price_for_type=info_carros.pivot_table(index="type", values="price", aggfunc="mean").reset_index().sort_values(by="price", ascending=False)
```

```python
price_for_type.plot(kind='bar', x="type", y="price")
```
